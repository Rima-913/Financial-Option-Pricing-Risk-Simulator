#### Ito Calculus

```python
# Simulate correlated Brownian motions
def simulate_correlated_paths(S0, mu, sigma, rho, T, dt, n_paths):
    """
    Simulate two correlated stock price paths using GBM and correlated Brownian motion
    """
    n_steps = int(T / dt)
    t = np.linspace(0, T, n_steps)


    cov = np.array([[1, rho], [rho, 1]])
    L = np.linalg.cholesky(cov)

    Z = np.random.normal(size=(2, n_steps, n_paths))


    W = np.einsum('ij,jkl->ikl', L, Z)


    S1 = np.zeros((n_steps, n_paths))
    S2 = np.zeros((n_steps, n_paths))
    S1[0] = S0[0]
    S2[0] = S0[1]

    for i in range(1, n_steps):
        S1[i] = S1[i-1] * np.exp((mu[0] - 0.5 * sigma[0]**2) * dt + sigma[0] * np.sqrt(dt) * W[0, i])
        S2[i] = S2[i-1] * np.exp((mu[1] - 0.5 * sigma[1]**2) * dt + sigma[1] * np.sqrt(dt) * W[1, i])

    return t, S1, S2
```

```python
#  infosys tcs
S0 = [1472.88,3820.96 ]
mu = [0.0867, -0.464]
sigma = [0.2339, 0.2092]
rho = 0.68
T = 1
dt = 1/252
n_paths = 1

t, S1, S2 = simulate_correlated_paths(S0, mu, sigma, rho, T, dt, n_paths)


for i in range(n_paths):
    plt.plot(t, S1[:, i], color='blue', alpha=0.5)
    plt.plot(t, S2[:, i], color='red', alpha=0.5)
plt.title('Simulated Stock Paths (Infosys and TCS)')
plt.xlabel('Time (Years)')
plt.ylabel('Stock Price')
plt.show()
```

Epps Effect

```python

def epps_effect_daily_data(log_returns1, log_returns2, freqs):
    correlations = []
    for f in freqs:

        sampled1 = log_returns1[::f]
        sampled2 = log_returns2[::f]


        returns1 = np.diff(sampled1)
        returns2 = np.diff(sampled2)

        if len(returns1) > 1 and len(returns2) > 1:
            corr = np.corrcoef(returns1, returns2)[0, 1]
            correlations.append(corr)
        else:
            correlations.append(np.nan)
    return correlations
```

```python

# Define frequencies (in days)
freqs = [1, 2, 5, 10, 20, 60]

correlations = epps_effect_daily_data(log_returns['INFY'], log_returns['TCS'], freqs)


plt.figure(figsize=(8,5))
plt.plot(freqs, correlations, marker='o', color='blue')
plt.xlabel('Sampling Interval (days)')
plt.ylabel('Measured Correlation')
plt.title('Correlation vs Sampling Frequency (INFY & TCS)')
plt.xscale('log')
# plt.gca().invert_xaxis()
plt.grid(False)
plt.show()
```
